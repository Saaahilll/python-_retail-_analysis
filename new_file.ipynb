```python
df.groupby(['gender'], as_index=False)['price'].sum().sort_values(by='price', ascending=False)
```

```python
df.groupby(['gender'], as_index=False)['price'].sum().sort_values(by='price', ascending=False)
sns.countplot(x= 'gender', data=df)
plt.title('sales_generated')
plt.show()
```

From above graph we can see that most purchase is done by males as compared to the othertwo showing their high purchasing power

AGE

```python
sns.countplot(x= 'age_bracket', hue= 'gender', data=df)
plt.title('Age Group')
plt.show()
```

Females spent the most in senior citizen category while Males dominate in the adults and Young age bracket

```python
sales_age= df.groupby(['age_bracket'],as_index=False)['price'].sum().sort_values(by='price', ascending=False)
sns.barplot(x='age_bracket', y='price' ,data=sales_age)
plt.title('sales_age')
plt.show()
```

```python
df.columns
```

```python
category_sales= df.groupby(['category_name'],as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False)
sns.barplot(x='category_name', y='quantity' ,data=category_sales)
sns.set(rc={'figure.figsize':(16,5)})
plt.title('category_sales')
plt.show()
```

Electronics made the most sales followed by sports& outdoors and fashion.

```python
payments_method = df.groupby(['payment_method', 'gender'], as_index=False).size()
sns.set(rc={'figure.figsize': (15, 5)})
ax = sns.barplot(x='payment_method', y='size', hue='gender', data=payments_method)
for p in ax.patches:
    ax.text(
        p.get_x() + p.get_width() / 2,  # x-coordinate (center of the bar)
        p.get_height(),                # y-coordinate (height of the bar)
        int(p.get_height()),           # Text (count value)
        ha='center', va='bottom'       # Alignment
    )
plt.title('Payments Methods by Gender')
plt.show()
```
